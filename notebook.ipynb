## Task 1

The given data contains **1500** rows and **8** columns. Firstly, the data needs to be validated according the the given criteria in the instructions. Looking at the columns one by one:

-  **rating** contained 150 ratings with NA value. Replaced those with 0. Data type was text and was changed to whole number.
-  **repeat_purchase** contained 0 or 1 with no missing values.
